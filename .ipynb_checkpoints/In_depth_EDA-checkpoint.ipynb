```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import pymysql                        
from sqlalchemy import create_engine  

from getpass import getpass 
```

```python
password = getpass()
```

```python
#connecting to MySQL airline database:
connection_string = 'mysql+pymysql://root:'+password+'@localhost/airlines'
engine = create_engine(connection_string)
```

The main focus of this notebook will be visualization, I will use it to produce the tables necessary for quality visualization in Tableau. I want to focus on the factors that makes an airline better meaning with higher overall rating compare to the others.

```python
airlines_df = pd.read_sql_query('SELECT * FROM airlines.airlines', engine)
airlines_df
```

```python
#looking at the total number of airlines
airlines_df['airline_name'].value_counts(dropna=False)
```

We can see that there is a total of 510 airlines with a maximum number of reviews of 100. I'll consider that airlines with a number of reviews < 100 are not suitable for an in-depth analysis. From those airlines I then want to select the top15 based on average overall rating.

```python
#Define SQL query:
query = '''SELECT airline_name, avg(overall_rating) AS avg_overall_rating, COUNT(airline_name) AS number_of_reviews FROM airlines.airlines
GROUP BY airline_name
HAVING COUNT(airline_name) =100
ORDER BY avg(overall_rating) DESC
LIMIT 15;'''

# Run SQL query and load data into a Pandas DataFrame
top_df = pd.read_sql_query(query, engine)
top_df
#majority of asian airlines are included in the top15
```

```python
name=top_df['airline_name'].tolist()
airlines_df=airlines_df[airlines_df['airline_name'].isin(name)]
airlines_df
```

For visualization purposes I want to add a column country based on the departure place of the reviews. Indeed, customers will have most of their interactions with the airline company's staff at the departure airport (considering that the boarding process is done by the airline's crew) rather than the arrival airport.

```python
#exporting top_airlines to SQL database:
airlines_df.to_sql('top15_airlines', con=engine, if_exists='replace', index=False)
```

```python
#adding the country column with a join (only keeping rows with a valid 'Country' value):
query_2= '''SELECT t.airline_name, t.overall_rating, t.verified, t.aircraft, t.type_of_traveller, t.seat_type, t.seat_comfort,
    t.cabin_staff_service, t.food_beverages, t.ground_service, t.inflight_entertainment, t.wifi_connectivity,
    t.value_for_money, t.recommended, t.time_before_review, t.delay_cancellation, t.flight_date, t.review_date, c.country 
FROM airlines.top15_airlines AS t
JOIN airlines.cities AS c 
ON t.departure_place=c.city_ascii;'''
airline_country = pd.read_sql_query(query_2, engine)
airline_country
```

37 rows were lost due to invalid/unrecognized departure place. The rest of the EDA will be done in Tableau using the last dataframe as a source table.

* Formatting the data in correct types for easy use in Tableau

```python
airline_country.info()
```

```python
#setting 'verified' and 'delay_cancellation' as boolean value:
airline_country['verified']=airline_country['verified'].astype('bool')
airline_country['delay_cancellation']=airline_country['delay_cancellation'].astype('bool')
airline_country
```

```python
#setting review date as date type and dropping flight date to keep only one time variable:
airline_country['review_date'] = pd.to_datetime(airline_country['review_date'], format = '%B %Y')
airline_country = airline_country.drop(['flight_date'], axis=1)
airline_country
```

```python
airline_country.to_csv('data/top_airlines.csv', index=False)
```
